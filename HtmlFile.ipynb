```python
import requests
import pandas as pd
from io import StringIO

url = "https://en.wikipedia.org/wiki/Animal"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)
response.raise_for_status()   # shows error if request fails

html = response.text

# Read tables from HTML string
tables = pd.read_html(StringIO(html),match="Phylum")

df = tables[0]
```
