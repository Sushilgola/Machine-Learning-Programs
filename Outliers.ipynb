```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# define dataset 
dataset = [12,13,14,14,15,17,17,19,102,12,15,15,14,16,16,105,19,10,18,17,16,107]
```

```python
plt.hist(dataset) # to print the histogram of data
```

```python
# Z score method
outliers = []
def detect_outlier(data):
    threshold = 3  # 3rd standard deviation
    mean = np.mean(data)
    std = np.std(data)

    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
    
```

```python
detect_outlier(dataset)
```

# IQR 
1. sort data 
2. calculate Q1(25%) and Q3(75%)
3. Find the lower fence(Q1-1.5*IQR)
4. Find the higher fence(Q3+1.5*IQR)

```python
# sort
dataset = sorted(dataset)
```

```python
dataset
```

```python
q1 = np.percentile(dataset,25)
q3 = np.percentile(dataset, 75)
q1,q3
```

```python
iqr = q3-q1
```

```python
iqr
```

```python
# find the lower fence 
lower_fence = q1-(1.5*iqr)
# higher fence 
higher_fence = q3+(1.5*iqr)
print(lower_fence, higher_fence)
```

```python
dataset
```

```python
out = []
def find_outliers(data):
    for i in data:
        if i>higher_fence or i<lower_fence:
            out.append(i)
    return out
```

```python
find_outliers(dataset)
```
