```python
days = ['mon', 'tue', 'web', 'thu', 'fri', 'sat','sun']
sales = [12,10,20,22,19,25,9]
plt.title('Grocery Shop Sales')
plt.plot(days, sales, color='red',linestyle=':', linewidth=2, marker='o', label='Grocery Shop')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.legend()
plt.grid(color='black', linestyle='--')
plt.xticks([1,2,3,4,5,6],['mon','tus','wed','thur','fri','sat'])
plt.show()
```

```python
import matplotlib.pyplot as plt

product = ['a', 'b', 'c', 'd', 'e']
sales = [1200,1000,1300,800,400]
plt.bar(product, sales, color='red', label='Shop Data Bar')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Shop Data')
# plt.label('Shop Data')
plt.show
```

```python
import matplotlib.pyplot as plt
u_data = ['CS', 'Bio', 'Math', 'Static', 'geology']
n_stu = [150, 180, 300, 200, 90]
plt.bar(u_data, n_stu, color='blue',label='University Bar')
# plt.plot(u_data, n_stu, color='blue',linestyle='--', label='University graph')
plt.xlabel('University Data')
plt.ylabel('Student Data')
plt.show()
```

```python
import matplotlib.pyplot as plt
direction = ['East','West','North','South']
revenue = [2000,3000,4500,1500]
plt.pie(revenue, labels=direction, autopct='%1.1f%%',colors=['green','red','blue','orange'])
plt.title('Region data')
plt.show()
```

# Scatter PLot

```python
import matplotlib.pyplot as plt
h_studies = [1,2,4,5,6]
exam_score = [60,68,70,89,95]
# Sctter Plot
plt.scatter(h_studies, exam_score, color='green',marker='o',linestyle=':',label='Student Exam')
plt.xlabel('Hour of Studies')
plt.ylabel('Exam Score')
plt.title('Student Exam Data')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt 
plt.scatter([1,2,3], [60,65,70], color='blue', marker='^', label='Class A')
plt.scatter([3,2,5], [65,69,80], color='red', marker='+', label='Class B')
plt.grid(True)
plt.title('Compare Two Class Data')
plt.show()
```

# Subplot Method 

```python
# plt.subplot(nrows,ncols, index)
import matplotlib.pyplot as plt
x = [0,2,4,6,8]
y = [1,3,5,3,9]
plt.subplot(1,2,1)
plt.plot(x,y, color='blue', label='Line')
plt.title('Line Chart')

plt.subplot(1,2,2)
plt.bar(x,y, color='red', label='Bar')
plt.title('Bar Chart')

plt.suptitle('Compare Two Shops Data')
plt.tight_layout()
plt.show()
```

# Savefig Function

```python
# savefig('filename.extension, dpi=
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,15,10,20,30]

# Create plot
plt.subplot(1,2,1)
plt.plot(x,y, color='red', label='Data',linestyle=':', marker='o')
plt.title('Line Graph Data')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.tight_layout()


plt.subplot(1,2,2)
plt.bar(x,y, color='yellow', label='Data',linestyle=':',)
plt.title('Line Graph Data')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.tight_layout()
plt.savefig('graph.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
