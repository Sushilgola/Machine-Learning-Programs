```python
# Find mean median and standard deviation of a given dataset
data = np.array([1,2,5,7,12,10,14,17,10,15,18,20])
df_mean = np.mean(data)
df_median = np.median(data)
df_std = np.std(data)
```

```python
print("mean ", df_mean)
print("median ", df_median)
print("Standard deviation ", df_std)
```

```python
# perform matrix addition and multiplicaiton on two arrays
arr1 = np.array([[2,4,6],
                [0,4,8]])
arr2 = np.array([[1,3,5],
                [1,7,5]])
sum_arr = np.add(arr1, arr2)
print("sum ", sum_arr)
mul_arr = np.multiply(arr1, arr2)
print("Multiply ", mul_arr)
```

```python
new_arr = data.reshape(3,4)
max_num = np.max(new_arr)
min_num = np.min(new_arr)
print(new_arr)
print("Maximum ", max_num)   
print("minimum ", min_num)
```

```python
# find row and column wise sum of array
row_wise = np.sum(new_arr, axis=1)
print(row_wise)
column_wise = np.sum(new_arr, axis=0)
print(column_wise)
```

```python
# Normalization of the array 
arr3 = np.array([2,3,5,6,7,8,9])
normalize = (arr3 - np.min(arr3)) / (np.max(arr3) - np.min(arr3))
print(normalize)
```
