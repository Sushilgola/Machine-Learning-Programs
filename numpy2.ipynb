# Modification of Array

```python
import numpy as np
```

```python
arr_2d = np.array([[1,2,3],
                   [3,4,5]])
print(arr_2d.shape)   # define the size of array
```

```python
print(arr_2d.size)  # count number of element in array
```

```python
arr1 = np.array([[[2,3],[3,4],[5,6]]])  # 3 dimension array
```

```python
print(arr_2d.ndim) # present the dimension of array
print(arr1.ndim)  # present 3d array
```

```python
arr2 = np.array([2.3,4.5,9,1])
print(arr2.dtype)   # represent the type of array 
```

```python
arr3 = np.array([3.6,2.9,1.9,2.7])
int_arr3 = arr3.astype(int)# convert any type of element in array
str_arr3 = arr3.astype(str)  # this make array in string type
print(int_arr3)
print(int_arr3.dtype)
print(str_arr3.dtype)
```

# Airthmatic operation in Array

```python
arr4 = np.array([10,20,30])
print(arr4+10)
print(arr4/5)
print(arr4-10)
print(arr4 ** 2)
```

# Aggregation function

```python
arr5 = np.array([[2,3,4],[4,6,8]])
print(np.sum(arr5))
print(np.mean(arr5))
print(np.min(arr5))
print(np.max(arr5))
print(np.std(arr5))   # find standard deviation
print(np.var(arr5))  # find varience
```

# Indexing in numpy

```python
import numpy as np
```

```python
arr = np.array([10,20,30,40,50,60])
print(arr) 
print(arr.dtype) # what kind of value store in array
print(arr[0]) 
print(arr[4])
print(arr[-5])
```

# Slicing in Array

```python
arr = np.array([12,23,14,17,19,20,26,28,25])
print(arr[0:5])
print(arr[-1:-6:-1])
print(arr[-1:-1-(len(arr)):-1])  # reverse the string
print(arr[1:6+1:])
```

# Fancy indexing

```python
arr = np.array([12,23,14,17,19,20,26,28,25]) 
print(arr[[1,3,5,7]])
```

# Filtering data in array

```python
arr = np.array([12,23,14,17,19,20,26,28,25])
print(arr[arr <= 20])
```

# Reshaping and manupulating array

```python
arr = np.array([12,23,14,17,19,20,26,28,25,22,23,29])
reshape_arr = arr.reshape(3,4)
reshape_arr
```

# Flattering  array

```python
arr = np.array([[12,23,14],[17,19,20],[26,28,25]])
print(arr.ravel())
print(arr.flatten())
```
