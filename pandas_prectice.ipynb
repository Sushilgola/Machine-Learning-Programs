```python
import pandas as pd
```

```python
# this is dictionary format
df = pd.DataFrame({'days':['mon','Tus','Wed','Thur'],
                  'temprature':[28,27,25,20],
                  'events':['Rain','Cloud','Normal','Cold'],
                  'windSpeed':[8,7,9,10]})
```
